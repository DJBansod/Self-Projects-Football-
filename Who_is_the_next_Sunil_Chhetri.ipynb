```python
indian_forwards = deepcopy(isl_df[(isl_df['position'] == 'Forward') & (isl_df['country_name'] == 'India')])
```

```python
indian_forwards.reset_index(inplace=True, drop=True)
```

```python
#Levelling the playing field
```

```python
fwd_mins_df = indian_forwards.groupby(['id', 'name'])['minutes_played'].sum().reset_index()
```

```python
px.bar(x= 'name', y= 'minutes_played', text= 'minutes_played', data_frame= fwd_mins_df)
```

```python
indian_forwards['goalsPer90'] = indian_forwards['events.goals'].divide(indian_forwards['actual_minutes_played']).multiply(90)
```

```python
indian_forwards = indian_forwards[indian_forwards['actual_minutes_played'] > 0]
```

```python
per90Cols = ['events.goals', 'events.assists', 'events.shots', 'events.shots_on_target', 'events.chances_created', 'events.fouls_suffered', 'touches.total', 'touches.aerial_duel.won', 'touches.ground_duel.won']
for col in per90Cols:
  indian_forwards[col + 'Per90'] = indian_forwards[col].divide(indian_forwards['actual_minutes_played']).multiply(90)
```

```python
#using List Comprehension
cols_for_radar = [i + 'Per90' for i in per90Cols]
print(cols_for_radar)
```

```python
indian_forwards.loc[(indian_forwards['name'] == 'Sunil Chhetri'), cols_for_radar].sum()
```

```python
#Analyzing Chettri's Performance
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
indian_forwards[cols_for_radar] = scaler.fit_transform(indian_forwards[cols_for_radar])
```

```python
fig = px.line_polar(indian_forwards, r = indian_forwards.loc[(indian_forwards['name'] == 'Sunil Chhetri'), cols_for_radar].sum(), theta = cols_for_radar, line_close= True)
fig.update_traces(fill = "toself")
fig.show()
```

```python
#Comparing with the other forwards
```

```python
for i, row in indian_forwards.iterrows():
  print(row['name'])
  fig = px.line_polar(indian_forwards, r = indian_forwards.loc[(indian_forwards['name'] == row['name']), cols_for_radar].sum(), theta = cols_for_radar, line_close= True, title= row['name'])
  fig.update_traces(fill = "toself")
  fig.show()
```

```python
#Replacement
```

```python
ind_fwd_names = indian_forwards.drop_duplicates(subset=['name'])[['id', 'name']]
```

```python
ind_fwd_names.shape
```

```python
ind_fwd_names.head()
```

```python
isl_max = indian_forwards[cols_for_radar].max().max()
```

```python
import plotly.graph_objects as go
```

```python
type(indian_forwards.loc[indian_forwards['name'] == 'Sunil Chettri'])
```

```python
for i, row in ind_fwd_names.iterrows():
  if row['id'] == 19150:
    continue
  print(row['name'])
  #Add Radar Plots for different players
  fig = go.Figure()   #to initiate the Plotly go figure 
  fig.add_trace(go.Scatterpolar(r = indian_forwards.loc[indian_forwards['id'] == 19150, cols_for_radar].sum(),
                                theta = cols_for_radar, fill = 'toself', name = 'Sunil Chettri'))
  fig.add_trace(go.Scatterpolar(r = indian_forwards.loc[(indian_forwards['name'] == row['name']), cols_for_radar].sum(), 
                                theta = cols_for_radar, fill = 'toself', name = row['name']))
  fig.update_layout(title = "Sunil Chettri vs " + row['name'],
                    polar = dict(radialaxis = dict(visible = True, range = [0, isl_max])), showlegend = True)
  fig.show()              
                    

```

**Conclusion**

From the figures we can draw two conclusions:

1) Leon Augustin and Liston Colaco are the two players which can replace Sunil Chettri in a period of 6 months or a year as they have the same good quality almost all the stats and even have greater quality in some of the stats which indicates they can be more or less ready to fill the role.

2) Devendra Murgaonkar and Rahim Ali are two young prospects which can be developed in years to come and can be deployed in Sunil Chettri's role as the stats figure indicates they can up their stats if more minutes are given to them.
